```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Example data
x = np.array([3, 2.5, 2])
y = np.array(ests)  # your actual y values
yerr = np.array([1.6833, 1.6833, 1.6833])

# Linear function
def linear(x, m, b):
    return m * x + b

# Weighted fit
popt, pcov = curve_fit(linear, x, y, sigma=yerr, absolute_sigma=True)
m, b = popt
m_err, b_err = np.sqrt(np.diag(pcov))

print(f"Best fit: y = {m:.4f}x + {b:.4f}")
```
