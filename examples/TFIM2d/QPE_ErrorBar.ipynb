```python
import matplotlib.pyplot as plt
import numpy as np

# Unit circle
theta = np.linspace(0, 2 * np.pi, 500)
x = np.cos(theta)
y = np.sin(theta)

# First point
ph1 = (-0.18 + 0.359j)
std1 = 0.02877853172
x1, y1 = ph1.real, ph1.imag

# Second point
ph2 = ph1/np.abs(ph1)
std2 = std1/np.abs(ph1)
x2, y2 = ph2.real, ph2.imag

# Create plot
plt.figure(figsize=(6, 6))

# Plot unit circle
plt.plot(x, y, label='Unit Circle', color='blue')

# Plot first point with error bars
plt.errorbar(x1, y1, xerr=std1, yerr=std1, fmt='o', color='red',
             ecolor='red', elinewidth=1.5, capsize=4, label='phase ± std')

# Plot second point with error bars
plt.errorbar(x2, y2, xerr=std2, yerr=std2, fmt='o', color='green',
             ecolor='green', elinewidth=1.5, capsize=4, label='phase projected ± amplified std')

# Dotted orange lines
plt.plot([0, x1], [0, y1], 'o--', color='orange', linewidth=1.2)
plt.plot([0, x2], [0, y2], 'o--', color='orange', linewidth=1.2)
plt.plot([x1, x2], [y1, y2], 'o--', color='orange', linewidth=1.2)

# Styling
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.legend()

# Limit to top-left quadrant: +Y and -X
plt.xlim(-1.1, 0.05)
plt.ylim(-0.05, 1.1)
plt.xlabel("Real Part")
plt.ylabel("Imag Part")

plt.show()
```
