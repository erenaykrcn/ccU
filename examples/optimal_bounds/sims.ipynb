```python
import numpy as np
import qiskit
from qiskit.quantum_info import state_fidelity
from numpy import linalg as LA
import qib
import matplotlib.pyplot as plt
import scipy
import h5py

X = np.array([[0, 1], [1, 0]])
Z = np.array([[1, 0], [0, -1]])
Y = np.array([[0, -1j], [1j, 0]])
XX = np.kron(X, X)
XI = np.kron(X, np.eye(2))
YY = np.kron(Y, Y)

def random_unitary():
    theta1, theta2 = (np.random.uniform(0, 2*np.pi), np.random.uniform(0, 2*np.pi))
    return scipy.linalg.expm(-1j*XX*theta1)@scipy.linalg.expm(-1j*YY*theta2)

def random_unitary_4():
    Z = (np.random.randn(4,4) + 1j*np.random.randn(4,4)) / np.sqrt(2)
    Q, R = np.linalg.qr(Z)
    diag = np.diag(R)
    lambdas = diag / np.abs(diag)
    Q = Q * lambdas
    return Q

def ansatz(theta):
    return scipy.linalg.expm(-1j*XI*theta[2])@scipy.linalg.expm(-1j*XX*theta[0])@scipy.linalg.expm(-1j*YY*theta[1])

class Cost():
    def __init__(self):
        self.G = random_unitary_4()
    def f(self, theta):
        return -np.trace(ansatz(theta)@self.G).real

cost = Cost()
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.linspace(0, 2*np.pi, 100)
xv, yv = np.meshgrid(x, y)

costs = np.zeros((100, 100))
for i in range(100):
    for j in range(100):
        costs[i, j] = cost.f([xv[i, j], yv[i, j]])
```
